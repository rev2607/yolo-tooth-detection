```python
import os
import shutil

# Paths
images_train_path = '/content/dataset/images/train'
images_val_path = '/content/dataset/images/val'
labels_train_path = '/content/dataset/labels/train'
labels_val_path = '/content/dataset/labels/val'

# Destination
base_path = '/content/ToothNumber_TaskDataset'
os.makedirs(base_path, exist_ok=True)

# Create subdirectories for YOLO format
for split in ['train', 'val']:
    os.makedirs(f'{base_path}/{split}/images', exist_ok=True)
    os.makedirs(f'{base_path}/{split}/labels', exist_ok=True)

# Copy training images and labels
for img in os.listdir(images_train_path):
    shutil.copy(os.path.join(images_train_path, img), f'{base_path}/train/images/{img}')

for lbl in os.listdir(labels_train_path):
    shutil.copy(os.path.join(labels_train_path, lbl), f'{base_path}/train/labels/{lbl}')

# Copy validation images and labels
for img in os.listdir(images_val_path):
    shutil.copy(os.path.join(images_val_path, img), f'{base_path}/val/images/{img}')

for lbl in os.listdir(labels_val_path):
    shutil.copy(os.path.join(labels_val_path, lbl), f'{base_path}/val/labels/{lbl}')

# Move data.yaml to dataset root
if os.path.exists('/content/data.yaml'):
    shutil.move('/content/data.yaml', f'{base_path}/data.yaml')

print("✅ YOLO dataset prepared successfully at:", base_path)
```

```python
from ultralytics import YOLO

# Load pretrained YOLOv8 model
model = YOLO('yolov8n.pt')  # Use yolov8s.pt for better accuracy, yolov8m.pt for more power

# Train the model
model.train(
    data='/content/ToothNumber_TaskDataset/data.yaml',
    epochs=50,        # Start with 50 epochs (adjust later)
    imgsz=640,        # Image size (can increase to 1024 for dental X-rays)
    batch=16,         # Adjust based on GPU memory
    workers=2,
    patience=10       # Early stopping
)
```

```python
metrics = model.val()
print(metrics)
```

```python
from ultralytics import YOLO

# 1) pick a stronger backbone
# 'yolov8l.pt' (large) usually fits on T4 15GB with imgsz=1024 and batch ~8
# if OOM: drop to batch=6 or imgsz=896; if room: try yolov8x.pt with batch=6
model = YOLO('yolov8l.pt')

results = model.train(
    data='/content/ToothNumber_TaskDataset/data.yaml',
    epochs=150,          # train longer
    imgsz=1024,          # higher resolution helps small teeth
    batch=8,             # adjust if OOM
    device=0,
    workers=2,
    patience=25,         # early stop if plateaus

    # ---- turn OFF geometry-breaking augs (FDI is position-specific) ----
    fliplr=0.0,
    flipud=0.0,
    mosaic=0.0,
    mixup=0.0,
    copy_paste=0.0,

    # ---- gentle, X-ray-friendly augs ----
    degrees=5.0,
    translate=0.02,
    scale=0.10,
    shear=0.0,
    perspective=0.0,
    hsv_h=0.0, hsv_s=0.0, hsv_v=0.0,  # X-rays are grayscale; skip color jitter

    # ---- optimization tweaks ----
    lr0=0.003,          # a bit higher start lr than default
    lrf=0.1,            # cosine-like decay depth
    momentum=0.937,
    weight_decay=0.0005,
    warmup_epochs=3.0,
    overlap_mask=True,  # harmless for detect; keeps pipeline defaults tidy
    pretrained=True,
    cache=True,         # speed up epochs
    verbose=True
)

# Validate (no TTA flips to avoid quadrant swaps)
metrics = model.val()  # uses last/best weights from the run
print(metrics)
```
